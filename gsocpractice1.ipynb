```python
import matplotlib.pyplot as plt#plotting with matplotlib
```

```python
x = np.arange(0, 10, 0.2)
y_sin=np.sin(x)
y_cos=np.cos(x)
```

```python
plt.plot(x, y_sin,label='Sine',linewidth=4.0)
plt.plot(x, y_cos, label='Cos')
plt.plot(x, y_cos, label='Cos_2')

plt.xlabel('x label')
plt.ylabel('y label')
plt.title("Simple Plot")
plt.legend()
plt.show()
plt.close()
```

```python
x = np.arange(0, 10, 0.2)#working with mutliple plots
```

```python
y_sin=np.sin(x)
y_cos=np.cos(x)
```

```python
plt.subplot(1,2,1)
plt.plot(x,y_sin)
plt.title('Sine Plot')

plt.subplot(1,2,2)
plt.plot(x,y_cos)
plt.title('Cos Plot')
```
